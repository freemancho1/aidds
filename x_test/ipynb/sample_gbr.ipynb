```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error

# 당뇨병 데이터셋 로드
diabetes = load_diabetes()
X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.2, random_state=42)

# GradientBoostingRegressor 모델 생성
model = GradientBoostingRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 모델 학습 및 각 단계에서의 예측값 추적
predictions = []
for stage_prediction in model.staged_predict(X_test):
    predictions.append(mean_squared_error(y_test, stage_prediction))

# 각 단계에서의 MSE 값 출력
for i, mse in enumerate(predictions):
    print(f"트리 {i+1}개 사용 시 MSE: {mse:.4f}")
```
