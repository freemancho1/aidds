```python
from PIL import Image, ImageDraw

# 이미지 크기 설정
width = 800
height = 400

# 이미지 생성
image = Image.new("RGB", (width, height), "white")
draw = ImageDraw.Draw(image)

# 시작점과 방향 설정
start_point = (50, 50)
direction = (1, 1)  # x 방향, y 방향 (오른쪽 아래로 이동)

# 벽면 좌표 설정
top_left = (0, 0)
top_right = (width, 0)
bottom_left = (0, height)
bottom_right = (width, height)

# 이동 경로 그리기 함수
def draw_path(point):
    draw.rectangle([point, (point[0] + 1, point[1] + 1)], fill="black")

# 반사 함수
def reflect(point, direction):
    x, y = point
    dx, dy = direction
    
    # 벽면과의 충돌 감지
    if x <= 0 or x >= width or y <= 0 or y >= height:
        # x 축 벽면과의 충돌
        if x <= 0 or x >= width:
            dx *= -1
        # y 축 벽면과의 충돌
        if y <= 0 or y >= height:
            dy *= -1
    return (x + dx, y + dy), (dx, dy)

# 반사 경로 그리기
draw_path(start_point)
for _ in range(200):  # 반사 횟수
    start_point, direction = reflect(start_point, direction)
    draw_path(start_point)

# 10단위로 좌표 찍기
for x in range(0, width, 10):
    for y in range(0, height, 10):
        if x == 0 or x>= width or y == 0 or y >= height:
            draw.rectangle([x, y, x + 1, y + 1], fill="red")

# 이미지 저장
image.show()
```
